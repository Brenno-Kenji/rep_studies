## **Data Preparation**

```python
# drop redudant columns
df = df.drop(['Column1', 'Column2', 'Column3', 'Column4', 'status_id', 'status_published'], axis = 1)

# check for NaN values again
df.isna().sum()
```

```python
# split in X and y our dataset
X = df
y = df['status_type']
```

```python
# transform categorical features with label encoding
le = LabelEncoder()
X['status_type'] = le.fit_transform(X['status_type'])

y = le.transform(y)
```

```python
X.info()
```

```python
# feature scalling
cols = X.columns

ms = MinMaxScaler()
X = ms.fit_transform(X)

X = pd.DataFrame(X, columns = [cols])
```

## **Modeling**

```python
# K-means >> 2 clusters
k_means = KMeans(n_clusters = 2, random_state = 0)
k_means.fit(X)
```

```python
k_means.cluster_centers_
```

```python
k_means.inertia_
```

```python
# check the quality of weak classification
labels = k_means.labels_

# check how many of the samples were correctly labeled
correct_labels = sum(y == labels)

print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
```

```python
print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

## **Find optimal number of clusters**

```python
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2,random_state=0)

kmeans.fit(X)

labels = kmeans.labels_

# check how many of the samples were correctly labeled

correct_labels = sum(y == labels)

print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))

print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```
